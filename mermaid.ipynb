```python
from IPython.display import Image, display
import requests
import urllib.parse
import base64 # Import the base64 module

def display_mermaid(graph):
    graphbytes = graph.encode("ascii")
    base64_bytes = base64.b64encode(graphbytes)
    base64_string = base64_bytes.decode("ascii")
    url = "https://mermaid.ink/img/" + base64_string
    response = requests.get(url)
    display(Image(response.content))
```

```python
# Example usage
mermaid_code = """
graph TD
    A[Start] --> B{Is it?}
    B -->|Yes| C[OK]
    C --> D[Rethink]
    D --> B
    B ---->|No| E[End]
"""

display_mermaid(mermaid_code)
```

```python
mermaid_code = """
graph TD
    A[Start Reconciliation] --> B[Initialize Spark Session]
    B --> C[Read Input Files]

    C --> D[Fetch Left Feed Config]
    C --> E[Fetch Right Feed Config]

    D --> F[Read Left Dataset]
    E --> G[Read Right Dataset]

    F --> H[Assign IDs Left]
    G --> I[Assign IDs Right]

    H --> J[Add Reserved Columns Left]
    I --> K[Add Reserved Columns Right]

    J --> L[Add Match Columns Left]
    K --> M[Add Match Columns Right]

    L --> N[Apply Filters Left]
    M --> O[Apply Filters Right]

    N --> P[Route to Reconciliation Handler]
    O --> P

    P --> Q[Fresh Reconciliation]
    P --> R[Incremental Reconciliation]
    P --> S[Fresh Incremental Reconciliation]

    Q --> T[Join Datasets]
    R --> T
    S --> T

    T --> U[Match and Identify Mismatches]
    U --> V[Calculate Statistics]
    V --> W[Generate Results]

    W --> X[Write Results to NAS Drive]
    W --> Y[Write Matches and Breaks to Hive]
    W --> Z[Publish Audit Messages to Kafka]

    X --> AA[Log Start and End Times]
    Y --> AA
    Z --> AA

    AA --> AB[End Reconciliation]
"""

display_mermaid(mermaid_code)
```
